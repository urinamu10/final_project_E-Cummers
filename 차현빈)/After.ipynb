## 결측치 삭제 및 대체

#### products

```python
products = products.dropna(subset=['name', 'brand']) # subset을 사용하면 다른 컬럼에 결측치가 있어도 유지.
```

#### orders

```python
no_data = orders[orders['delivered_at'].notnull()]

status_counts = no_data['status'].value_counts()

print(status_counts)
```

```python
# returned_at가 값이 있는 경우 :  Returned(12581) / 값이 없는 경우 : Shipped(37763), Complete(31344), Processing(25287), Cancelled(18492)
# shipped_at 값이 있는 경우 : Shipped(37763), Complete(31344), Returned(12581) / 값이 없는 경우 : Processing(25287), Cancelled(18492)
# delivered_at 값이 있는 경우 : Complete(31344), Returned(12581) / 값이 없는 경우 : Shipped(37763), Processing(25287), Cancelled(18492)
```

```python
orders['returned_at'] = orders['returned_at'].fillna(0)
orders['shipped_at'] = orders['shipped_at'].fillna(0)
orders['delivered_at'] = orders['delivered_at'].fillna(0)
```

#### order_items

```python
no_data = order_items[order_items['delivered_at'].notnull()]

status_counts = no_data['status'].value_counts()

print(status_counts)
```

```python
# returned_at가 값이 있는 경우 :  Returned(18128) / 값이 없는 경우 : Shipped(54641), Complete(45127), Processing(36961), Cancelled(26891)
# shipped_at 값이 있는 경우 : Shipped(54641), Complete(45127), Returned(18128) / 값이 없는 경우 : Processing(36961), Cancelled(26891)
# delivered_at 값이 있는 경우 : Complete(45127), Returned(18128) / 값이 없는 경우 : Shipped(54641), Processing(36961), Cancelled(26891)
```

```python
order_items['returned_at'] = order_items['returned_at'].fillna(0)
order_items['shipped_at'] = order_items['shipped_at'].fillna(0)
order_items['delivered_at'] = order_items['delivered_at'].fillna(0)
```

#### inventory_items

```python
inventory_items['sold_at'] = inventory_items['sold_at'].fillna(0) # 매개변수 만들기?
```

## 추가 칼럼 수정

#### 1. España -> Spain

```python
users['country'].value_counts()
```

```python
# 1. 'Espana'를 'Spain'으로 변경합니다.
# dict 형태 {'기존값': '바꿀값'}를 사용하여 정확하게 매칭합니다.
users['country'] = users['country'].replace({'España': 'Spain'})

# 2. 결과 확인 (제대로 바뀌었는지 unique 값 확인)
print(users['country'].unique())
```

#### 2. id 칼럼 이름 변경

```python
# users
# id (고객 고유 ID) - user_id로 변경

# products
# id (상품 고유 ID) - product_id로 변경

# orders
# order_id (주문 고유 ID)
# user_id (주문한 고객 ID)

# order_items
# id (아이템별 고유 ID) - order_items_id로 변경
# order_id (주문 ID)
# user_id (유저 ID)
# product_id (상품 ID)
# inventory_item_id (연결된 재고 ID)

# inventory_items
# id (재고 고유 ID) - inventory_items_id로 변경
# product_id (상품 ID)

# distribution_centers
# id (센터 ID) - centers_id로 변경

# events
# id (이벤트 고유 ID) - event_id로 변경
# user_id (유저 ID)
# session_id (세션 ID)
```

```python
users = users.rename(columns={'id': 'user_id'})

print("변경된 칼럼 목록:", users.columns.tolist())
```

```python
products = products.rename(columns={'id': 'product_id'})

print("변경된 칼럼 목록:", products.columns.tolist())
```

```python
order_items = order_items.rename(columns={'id': 'order_item_id'})

print("변경된 칼럼 목록:", order_items.columns.tolist())
```

```python
inventory_items = inventory_items.rename(columns={'id': 'inventory_item_id'})

print("변경된 칼럼 목록:", inventory_items.columns.tolist())
```

```python
distribution_centers = distribution_centers.rename(columns={'id': 'center_id'})

print("변경된 칼럼 목록:", distribution_centers.columns.tolist())
```

```python
events = events.rename(columns={'id': 'event_id'})

print("변경된 칼럼 목록:", events.columns.tolist())
```
