#### users

```python
users[users['city'].isnull()]
```

```python
# 1. 비교를 위해 위도/경도를 소수점 2자리로 반올림한 값을 만듭니다.
# (소수점 2자리는 약 1.1km 차이로, 같은 '동네'를 의미합니다)
users['lat_round'] = users['latitude'].round(2)
users['long_round'] = users['longitude'].round(2)

# 2. 도시 이름이 있는 데이터로 '좌표-도시' 정답지(사전)를 만듭니다.
geo_dict = users.dropna(subset=['city']).set_index(['lat_round', 'long_round'])['city'].to_dict()

# 3. 도시가 없는 데이터들이 이 정답지에서 몇 개나 이름을 찾을 수 있는지 확인합니다.
null_users = users[users['city'].isnull()]
# 좌표 쌍을 튜플로 묶어서 정답지에 있는지 대조
found_count = null_users.apply(lambda x: (x['lat_round'], x['long_round']) in geo_dict, axis=1).sum()

print(f"도시가 없는 유저 {len(null_users)}명 중,")
print(f"근처 좌표를 통해 도시를 찾을 수 있는 유저: {found_count}명")
```

#### events

```python
# 1. 세션별로 user_id가 몇 종류(개수) 있는지 계산합니다.
# (로그인 전후가 섞여있다면 user_id 종류가 1개 이상일 것입니다)
session_user_counts = events.groupby('session_id')['user_id'].nunique()

# 2. 그중에서 user_id가 1개 이상 발견된 세션 아이디만 가져옵니다.
identified_sessions = session_user_counts[session_user_counts > 0].index

# 3. 위에서 찾은 세션들 중에서, 여전히 user_id가 결측치(NaN)인 데이터만 출력합니다.
# 이 데이터는 나중에 로그인한 '누구'의 익명 시절 기록이구나!"라고 알 수 있습니다.
result = events[events['session_id'].isin(identified_sessions) & events['user_id'].isnull()]

result

# 값이 없음 즉, 비회원 사람들은 로그인 한적이 한번도 없다
```

```python
# 비회원만 따로 봐야 할 듯
```

```python
# 1. user_id는 있는데 city가 비어있는 '복구 가능' 행 개수
recoverable = events[events['user_id'].notnull() & events['city'].isnull()]
print(f"회원 정보로 복구 가능한 이벤트 수: {len(recoverable)}")

# 2. user_id도 없고 city도 없는 행 개수
mysterious = events[events['user_id'].isnull() & events['city'].isnull()]
print(f"비회원이라 추적이 어려운 이벤트 수: {len(mysterious)}")
```

```python
# 1. 조건 필터링: city가 결측치(isnull)이면서, user_id는 결측치가 아닌(notnull) 행만 추출
recoverable_events = events[events['city'].isnull() & events['user_id'].notnull()].copy()

# 2. 정렬: user_id별로 묶어서 보고, 그 안에서는 시간(created_at / 페이지를 조회한 시각) 순으로 정렬
sorted_recoverable = recoverable_events.sort_values(by=['user_id', 'created_at'])

sorted_recoverable
```

```python
# 1. 고유한 유저 명단을 따로 저장합니다.
target_user_ids = sorted_recoverable['user_id'].unique()

# 2. .isin()을 사용하여 '명단에 포함된 유저'이면서 '도시 정보가 있는' 행만 필터링합니다.
correct_result = events[events['user_id'].isin(target_user_ids) & events['city'].notnull()]

# 결과 확인
correct_result

# 대체할 수 없음
```

## 시각화

```python
plt.figure(figsize=(10, 6))

# 히스토그램 그리기
sns.histplot(users['age'], bins=30, kde=True, color='skyblue')

plt.title('theLook 고객 연령대 분포', size=15)
plt.xlabel('나이')
plt.ylabel('고객 수')
plt.axvspan(20, 39, color='yellow', alpha=0.3, label='2030 타겟존') # 2030 강조
plt.legend()
plt.show()
```

```python
# 1. 데이터 합치기 (이름과 판매량을 묶습니다)
merged = pd.merge(order_items, products, left_on='product_id', right_on='id')

# 3. 상품명별로 몇 개 팔렸는지 딱 10개만 추리기
top10 = merged['name'].value_counts().head(10)

# 4. 그리기
plt.figure(figsize=(10, 6))
top10.plot(kind='barh', color='royalblue') # 가로 바 차트

# 5. 마무리
plt.title('theLook 전세계 인기 상품 TOP 10', size=15)
plt.xlabel('판매 개수')
plt.gca().invert_yaxis() # 높은 순서대로 위에서부터 나오게 뒤집기
plt.show()
```

```python
# 2. 데이터 병합 (주문 + 유저 + 상품)
merged_all = pd.merge(order_items, users[['id', 'age']], left_on='user_id', right_on='id')
merged_all = pd.merge(merged_all, products[['id', 'category']], left_on='product_id', right_on='id')

# 3. 2030 유저 데이터만 추출
target_2030 = merged_all[(merged_all['age'] >= 20) & (merged_all['age'] <= 39)]

# 4. 카테고리별 판매량 집계
category_2030 = target_2030['category'].value_counts().head(10)

# 5. 시각화 (세련된 파스텔 톤 적용)
plt.figure(figsize=(12, 8))
sns.barplot(x=category_2030.values, y=category_2030.index, palette='coolwarm')

plt.title('2030 세대가 가장 많이 구매한 카테고리 TOP 10', size=18, pad=20)
plt.xlabel('총 판매 건수')
plt.ylabel('카테고리 명')

# 막대 끝에 수치 표시
for i, v in enumerate(category_2030.values):
    plt.text(v + 3, i, f'{v:,}건', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 2. 데이터 병합 (주문 + 유저 + 상품)
# 필요한 컬럼만 선택해서 병합하면 속도가 빠릅니다.
merged_brands = pd.merge(order_items[['user_id', 'product_id']], users[['id', 'age']], left_on='user_id', right_on='id')
merged_brands = pd.merge(merged_brands, products[['id', 'brand']], left_on='product_id', right_on='id')

# 3. 2030 유저 필터링
target_2030 = merged_brands[(merged_brands['age'] >= 20) & (merged_brands['age'] <= 39)]

# 4. 브랜드별 판매량 집계 및 상위 10개 추출
top_brands_2030 = target_2030['brand'].value_counts().head(10)

# 5. 시각화
plt.figure(figsize=(12, 8))
# 색상 팔레트를 'rocket'으로 설정하여 세련된 느낌을 줍니다.
sns.barplot(x=top_brands_2030.values, y=top_brands_2030.index, palette='rocket')

plt.title('2030 세대가 가장 많이 선택한 브랜드 TOP 10', size=18, pad=20)
plt.xlabel('판매량 (건)')
plt.ylabel('브랜드명')

# 데이터 레이블 추가
for i, v in enumerate(top_brands_2030.values):
    plt.text(v + 1, i, f'{v:,}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# 2. 데이터 결합 (유저 + 주문 + 상품)
# 유저의 연령 정보를 주문 데이터와 연결해야 합니다.
merged_2030 = pd.merge(order_items, users[['id', 'age']], left_on='user_id', right_on='id')
merged_2030 = pd.merge(merged_2030, products[['id', 'name', 'category']], left_on='product_id', right_on='id')

# 3. 2030 세대 필터링 (20세 이상 39세 이하)
target_2030 = merged_2030[(merged_2030['age'] >= 20) & (merged_2030['age'] <= 39)]

# 4. 상품별 판매량 집계 및 상위 10개 추출
top_10_2030 = target_2030['name'].value_counts().head(10)

# 5. 시각화
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_2030.values, y=top_10_2030.index, palette='spring')

plt.title('2030 세대가 가장 많이 구매한 TOP 10 상품', size=18, pad=20)
plt.xlabel('판매량 (건)')
plt.ylabel('상품명')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
