```python
import pandas as pd
ar = pd.read_csv('articles.csv')
cus = pd.read_csv('customers.csv')
sam = pd.read_csv('sample_submission.csv')
```

```python
import pandas as pd
tra = pd.read_csv('transactions_train.csv')
```

```python
columns_to_drop = ['detail_desc','product_code','product_type_no','graphical_appearance_no','colour_group_code','perceived_colour_value_id','perceived_colour_master_id','department_no','index_code','index_group_no','section_no','garment_group_no']

art_df = ar.drop(columns=columns_to_drop)

columns_to_drop = ['postal_code']

custom_df = cus.drop(columns=columns_to_drop)

custom_df.loc[(custom_df['FN'].isna()) & (custom_df['fashion_news_frequency'] == 'Regularly'), 'FN'] = 1
custom_df.loc[(custom_df['FN'].isna()) & (custom_df['fashion_news_frequency'] == 'Monthly'), 'FN'] = 1
custom_df.loc[custom_df['fashion_news_frequency'].isna(), 'fashion_news_frequency'] = 'NONE'
custom_df.loc[custom_df['club_member_status'].isna(), 'club_member_status'] = 'unknown'
custom_df['FN'].fillna(0, inplace=True)
custom_df['Active'].fillna(0, inplace=True)
custom_df['club_member_status'].fillna('Unknown', inplace=True)
custom_df.dropna(subset=['age'], inplace=True)
def age_group(age):
    if age < 20: return '10대'
    elif age < 30: return '20대'
    elif age < 40: return '30대'
    elif age < 50: return '40대'
    elif age < 60: return '50대'
    elif age < 70: return '60대'
    elif age < 80: return '70대'
    else: return '80대'
custom_df['age_group'] = custom_df['age'].apply(age_group)

tra['price']=pd.to_numeric(tra['price'], errors='coerce')*590
tra['price'] = tra['price'].round(2)
tra['t_dat'] = pd.to_datetime(tra['t_dat'])
```
