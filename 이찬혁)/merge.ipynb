```python

final['t_dat'] = pd.to_datetime(final['t_dat'])
# 분석 기준일 설정 (데이터상 마지막 날의 다음날)
date_df = final['t_dat'].max() + pd.Timedelta(days=1)

# ---------------------------------------------------------
# 1. Recency (최근성): 마지막 구매일로부터 며칠이 지났는가
# ---------------------------------------------------------
rfm_r = final.groupby('customer_id')['t_dat'].max().reset_index()
rfm_r['Recency'] = (date_df - rfm_r['t_dat']).dt.days
# 필요한 컬럼만 남기기 (t_dat는 이제 필요 없음)
rfm_r = rfm_r[['customer_id', 'Recency']]

# ---------------------------------------------------------
# 2. Frequency (빈도): 방문한 날짜가 총 며칠인가 (중복 제거)
# ---------------------------------------------------------
rfm_f = final.groupby('customer_id')['t_dat'].nunique().reset_index()
rfm_f.columns = ['customer_id', 'Frequency']

# ---------------------------------------------------------
# 3. Monetary (금액): 총 구매 금액은 얼마인가
# ---------------------------------------------------------
rfm_m = final.groupby('customer_id')['price'].sum().reset_index()
rfm_m.columns = ['customer_id', 'Monetary']

# ---------------------------------------------------------
# 4. 데이터 합치기 (Merge)
# ---------------------------------------------------------
# R과 F를 먼저 합치고, 그 결과에 M을 합칩니다.
rfm = pd.merge(rfm_r, rfm_f, on='customer_id')
rfm = pd.merge(rfm, rfm_m, on='customer_id')

# 최종 결과 확인
print(rfm.head())
```

```python
# R, F, M의 최솟값, 최댓값, 평균, 25/50/75% 지점을 한눈에 보여줍니다.
print(rfm[['Recency', 'Frequency', 'Monetary']].describe())
#Recency (최근성): 마지막 구매 후 며칠 지났나?
#가장 최근에 온 사람 (min):(1일 전)
#딱 중간인 사람 (50%):150일(약 5개월 전)
#가장 오래된 사람 (max):734일(약 2년 전)

#Frequency(빈도)
# 대부분의 사람 (25%~50%): 1번에서 3번 
# 나름 단골 (75%): 8번 
# 최고 단골 (max): 427번

# Monetary (금액): 총 얼마를 썼나?
#평균 (mean): 약 386 
#딱 중간인 사람 (50%): 약 146 
#큰손 고객 (75% 이상): 417에서 최대 34,029까지 썼습니다
```

```python
# 1. Recency 점수 
def get_r_score(x):
    if x <= 49: return 5      # 상위 25% (매우 최근)
    elif x <= 150: return 4   # 상위 50% (보통)
    elif x <= 395: return 3   # 하위 50% (조금 오래됨)
    elif x <= 500: return 2   # 하위 25% (오래됨)
    else: return 1            # 최하위 (이탈 가능성 높음)

# 2. Frequency 점수 (클수록 좋음)
def get_f_score(x):
    if x >= 8: return 5       # 상위 25% (자주 방문)
    elif x >= 3: return 4      # 상위 50% (보통)
    elif x >= 2: return 3      # 하위 50% (가끔 방문)
    else: return 1             # 1회 방문 (신규 또는 단발성)
    # *참고: 데이터상 25% 지점이 1이라서 1점과 2점의 차이가 적어 범위를 조정함

# 3. Monetary 점수 (클수록 좋음)
def get_m_score(x):
    if x >= 417: return 5     # 상위 25% (큰손)
    elif x >= 145: return 4    # 상위 50% (중간)
    elif x <= 52: return 1     # 하위 25% (소액 구매)
    else: return 3             # 그 사이

# 점수 적용
rfm['R_score'] = rfm['Recency'].apply(get_r_score)
rfm['F_score'] = rfm['Frequency'].apply(get_f_score)
rfm['M_score'] = rfm['Monetary'].apply(get_m_score)

# 합산 점수 및 등급 부여
rfm['Total_Score'] = rfm['R_score'] + rfm['F_score'] + rfm['M_score']

# 결과 확인
print(rfm[['customer_id', 'R_score', 'F_score', 'M_score', 'Total_Score']].head())
```

```python
def segment(row):
    r, f, m = row['R_score'], row['F_score'], row['M_score']
    total = row['Total_Score']
    
    # 1. VIP (Champions): 모든 지표가 최상위
    if r >= 4 and f >= 4 and m >= 4:
        return 'VIP'
    # 2. 충성 고객 (Loyal Customers): 자주 오고 점수가 높은 우량 고객
    if total >= 11:
        return '충성 고객'
    # 3. 신규 고객 (New Customers): 최근에 처음 방문 (R은 높지만 F는 낮음)
    if r == 5 and f < 4:
        return '신규고객'
    # 4. 잠재 충성 고객 (Potential Loyalists): 최근 방문 빈도가 높아지는 중
    if r >= 4 and f >= 3:
        return '최근 방문빈도가 높은 고객'
    # 5. 놓치면 안 될 큰손 (Can't Lose Them): 돈은 많이 썼으나 최근 소식이 없음
    if r <= 2 and m == 5:
        return '최근소식없는 큰손'
    # 6. 이탈 징후 (About to Sleep): 전반적인 점수가 낮아지는 중
    if total >= 6:
        return '이탈위험고객'
    # 7. 이탈 고객 (Lost): 모든 점수가 최하위권
    return '이탈고객'

# 데이터프레임의 각 행(row)에 함수 적용
rfm['Segment'] = rfm.apply(segment, axis=1)

# 결과 확인: 각 그룹에 몇 명씩 배정되었는지 확인
print("-- Segment Distribution --")
print(rfm['Segment'].value_counts())
print("-- Segment Ratio (%) --")
print(rfm['Segment'].value_counts(normalize=True) * 100)
```

```python
#코로나데이터 확인
import pandas as pd

# 1. 날짜 데이터 형식 변환 (이미 되어있다면 생략 가능)
# final['t_dat'] = pd.to_datetime(final['t_dat'])

# 2. 코로나 전(2018~2019)과 후(2020)를 구분하는 파생변수 생성
final['period'] = final['t_dat'].dt.year.apply(lambda x: 'Pre-Corona' if x < 2020 else 'Post-Corona')

# 3. 기간별 + 고객별 방문 빈도(Frequency) 계산
# 고객이 두 기간 모두 존재할 수 있으므로 두 기준으로 그룹화합니다.
rfm_comparison = final.groupby(['period', 'customer_id'])['t_dat'].nunique().reset_index()

# 컬럼명 변경 (t_dat -> frequency)
rfm_comparison.rename(columns={'t_dat': 'frequency'}, inplace=True)
```

```python
# 기간별 평균 방문 빈도 계산
analysis_result = rfm_comparison.groupby('period')['frequency'].mean()
analysis_result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=rfm_comparison, x='period', y='frequency', ci=None)
plt.title('Average Purchase Frequency: Pre vs Post Corona')
plt.ylabel('Average Visit Days per Customer')
plt.show()
```

```python
# 1. 각 기간의 실제 일수(Duration) 계산
# Pre-Corona: 2018-09-20 ~ 2019-12-31
# Post-Corona: 2020-01-01 ~ 2020-09-22

pre_days = (pd.to_datetime('2019-12-31') - pd.to_datetime('2018-09-20')).days
post_days = (pd.to_datetime('2020-09-22') - pd.to_datetime('2020-01-01')).days

print(f"Pre-Corona 실제 일수: {pre_days}일")
print(f"Post-Corona 실제 일수: {post_days}일")

# 2. 일평균 방문 빈도 계산 함수
def get_daily_freq(row):
    if row['period'] == 'Pre-Corona':
        return row['frequency'] / pre_days
    else:
        return row['frequency'] / post_days

rfm_comparison['daily_freq'] = rfm_comparison.apply(get_daily_freq, axis=1)

# 3. 결과 비교
final_analysis = rfm_comparison.groupby('period')['daily_freq'].mean()
print(final_analysis)
```

```python
# 1. 기간별 + 채널별 + 고객별 방문 빈도(Frequency) 계산
# nunique()를 사용하여 고객별/채널별 '방문 일수'를 구합니다.
channel_comparison = final.groupby(['period', 'sales_channel_id', 'customer_id'])['t_dat'].nunique().reset_index()
channel_comparison.rename(columns={'t_dat': 'frequency'}, inplace=True)

# 2. 일평균 방문 빈도 보정 (이미 계산된 pre_days, post_days 활용)
def get_daily_freq_channel(row):
    if row['period'] == 'Pre-Corona':
        return row['frequency'] / pre_days
    else:
        return row['frequency'] / post_days

channel_comparison['daily_freq'] = channel_comparison.apply(get_daily_freq_channel, axis=1)

# 3. 결과 비교: 기간별, 채널별 평균 일일 방문 빈도(1: 오프라인, 2: 온라인)
# 오프라인 코로나이전에비해 15.3%증가
# 온라인 코로나이전에비해 27%증가
channel_analysis = channel_comparison.groupby(['period', 'sales_channel_id'])['daily_freq'].mean().unstack()
print(channel_analysis)
```
