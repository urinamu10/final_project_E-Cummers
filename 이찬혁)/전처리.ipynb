```python
import pandas as pd
import numpy as np
event_df = pd.read_csv('events.csv')
event_df
```

```python
event_df.dtypes
```

```python
event_df.isna().sum()
```

```python
event_df['event_type'].value_counts().sort_index(ascending=True)
```

```python
#user_id가 결측치인유저의 행동(비회원인지 확인)
event_df[event_df['user_id'].isna()]['event_type'].value_counts()
```

```python
# user_id의 null 값을 0으로 채우기
event_df['user_id'] = event_df['user_id'].fillna(0)
# user_id가 0인값(비회원)따로 저장
non_member_df = event_df[event_df['user_id'] == 0]
non_member_df
```

```python
# 'city' 컬럼이 null인 행만 골라서 삭제
event_df = event_df.dropna(subset=['city'])
```
